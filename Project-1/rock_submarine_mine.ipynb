```python
import pandas as pd
import numpy as np
from sklearn.model_selection import train_test_split
from sklearn.linear_model import LogisticRegression
from sklearn.metrics import accuracy_score
```

```python
df=pd.read_csv('E:\ML-DL Projects\Project-1\sonar.csv',header=None)
```

```python
df.head()
```

```python
df.shape
```

```python
df.describe()
```

```python
df.isnull().sum()
```

```python
df[60].value_counts() # M-> mine R-> Rock
```

```python
df.groupby(60).mean()
```

```python
df[60]=df[60].replace({'R':1,'M':0})
```

```python
X=df.drop(columns=60,axis=1)
Y=df[60]
```

```python
Y
```

```python
X_train,X_test,Y_train,Y_test=train_test_split(X,Y,train_size=0.1,random_state=42,stratify=Y)
```

```python
print(X_train.shape,X_test.shape,Y_test.shape,Y_train.shape)
```

```python
model=LogisticRegression()
```

```python
print(X_train)
print(Y_train)
```

```python
model.fit(X_train,Y_train)
```

```python
#Acurracy score on traning data
X_train_prediction=model.predict(X_train)
Accuracy=accuracy_score(X_train_prediction,Y_train)
print(f"The Accuracy is {Accuracy}")
```

```python
#Acurracy score on testing data
X_test_prediction=model.predict(X_test)
Accuracy=accuracy_score(X_test_prediction,Y_test)
print(f"The Accuracy is {Accuracy}")
```
